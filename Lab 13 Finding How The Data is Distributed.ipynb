### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
pd.set_option('display.max_columns',None)
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
df.info(verbose=True,show_counts=True)
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
df['YearsCodePro'] = df['YearsCodePro'].replace({'Less than 1 year':0,'More than 50 years':51})
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
df['YearsCodePro'] = df['YearsCodePro'].fillna(df['YearsCodePro'].median())
df['RemoteWork'] = df['RemoteWork'].fillna(df['RemoteWork'].mode())
df['EdLevel'] = df['EdLevel'].fillna(df['EdLevel'].mode())
```

```python
df['Country'] = df['Country'].replace({'United States of America':'USA',
                                       'United Kingdom of Great Britain and Northern Ireland':'UK',
                                       'Republic of Korea':'South Korea',
                                       'Iran, Islamic Republic of...' : 'Iran',
                                       "Democratic People's Republic of Korea" : 'North Korea',
                                       'Venezuela, Bolivarian Republic of...':'Venezuela',
                                       'Micronesia, Federated States of...':'Micronesia',
                                      'Congo, Republic of the...':'Congo',
                                      'Democratic Republic of the Congo':'Congo'})
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
for i in ['Employment','JobSat','YearsCodePro']:
    print(df[i].value_counts())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
sns.kdeplot(data=df,x='JobSat')
plt.axvline(df['JobSat'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(df['JobSat'].median(), color='green', linestyle='-', label='Median')
plt.legend()
plt.show()

#Concentrated Satisfaction: There is a significant concentration of data points between the values of 7 and 9, with the highest peak (mode) appearing near 8. 
#This suggests that a large majority of respondents rate their job satisfaction quite highly.

#Multimodal Tendencies: The presence of multiple smaller peaks (around 4, 6, and 10) suggests that there are distinct sub-groups within the population. 
#While most are satisfied, there is a secondary cluster of moderately satisfied individuals near the 6 mark.

#Low Dissatisfaction: The density drops off sharply below 4, indicating that very few respondents reported low job satisfaction.

#Conclusion: Overall, the workforce represented in this data appears to have a high morale, as the "mass" of the probability density is heavily weighted 
#toward the upper end of the scale 
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
lang_worked =df['LanguageHaveWorkedWith'].str.split(';').explode()
lang_want = df['LanguageWantToWorkWith'].str.split(';').explode()

worked_counts = lang_worked.value_counts()
want_counts = lang_want.value_counts()

lang_comparison = pd.concat([worked_counts,want_counts],axis=1)
lang_comparison.columns = ['Have Worked With','Want to Work With']

top_10_langs = lang_comparison.sort_values(by='Have Worked With',ascending=False).head(10)

#plotting grouped bargraph
top_10_langs.plot(kind='bar', figsize=(12, 6))
plt.title('Top 10 Languages: Current vs. Desired')
plt.xlabel('Programming Language')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
c=df['Country'].value_counts().head(10).index
df_top_10=df[df['Country'].isin(c)]

plt.figure(figsize=(14, 7))
sns.countplot(data=df_top_10, x='Country', hue='RemoteWork')

plt.title('Remote Work Distribution by Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Respondents')
plt.legend(title='Work Type', loc='upper right')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
print(df[['JobSat','YearsCodePro']].corr())

from scipy import stats
df_clean = df[['JobSat', 'YearsCodePro']].dropna()
pearson_r,p_value = stats.pearsonr(df_clean['JobSat'],df_clean['YearsCodePro'])

print(f"\nPearson Correlation Coefficient: {pearson_r:.4f}")
print(f"P-value: {p_value:.4e}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
df['Employment']=df['Employment'].str.split(';').str[0]
df['EdLevel'] = df['EdLevel'].str.split('(').str[0]

ct = pd.crosstab(df['EdLevel'],df['Employment'])

ct.plot(kind='bar',stacked=True,figsize=(12, 6))
```

```python
sns.heatmap(ct, annot=True, fmt='d', cmap='YlGnBu')
```
