### Task 2: Analysis of Duplicate Response Patterns
1. Identify duplicate response patterns based on selected columns such as MainBranch, Employment, and RemoteWork.
2. Clarify that these represent multiple respondents with identical answers rather than duplicate records. Analyse which other columns frequently share identical values within these response-pattern groups.
   


```python
## Write your code here
import pandas as pd

# --- Part 1: Identify duplicate response patterns ---
# Group by the specified columns and count the number of respondents in each group
pattern_cols = ['MainBranch', 'Employment', 'RemoteWork']
response_patterns = df.groupby(pattern_cols).size().reset_index(name='Total Respondents')

# Sort to find the top 5 most common patterns
top_patterns = response_patterns.sort_values(by='Total Respondents', ascending=False).head(5)

print("Top 5 Shared Response Patterns:")
display(top_patterns)


# --- Part 2: Analyze which other columns share identical values ---
# Get the values of the largest group (the first row of our sorted top_patterns)
largest_group = top_patterns.iloc[0]

# Filter the original dataframe to ONLY include people from this largest group
largest_group_df = df[
    (df['MainBranch'] == largest_group['MainBranch']) &
    (df['Employment'] == largest_group['Employment']) &
    (df['RemoteWork'] == largest_group['RemoteWork'])
]

# Find other columns where everyone gave the exact same answer
# We do this by checking if the number of unique values (nunique) is exactly 1
identical_columns = []
for col in largest_group_df.columns:
    if col not in pattern_cols and largest_group_df[col].nunique() == 1:
        identical_columns.append(col)

print("-" * 50)
print(f"In the largest pattern group, there are {len(identical_columns)} other columns where everyone's answer is identical.")
print(f"Some of those column names are: {identical_columns}")
```

### Task 3: Visualize Shared Response Patterns
1. Create visualizations to show the distribution of shared response patterns across different categories.
2. Use bar charts or pie charts to represent the distribution of respondents who share identical values for MainBranch, Employment, and RemoteWork, grouped by Country and Employment.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns 

grouped_data = largest_group_df.groupby(['Country', 'Employment']).size().reset_index(name='Count')

top_10_groups = grouped_data.sort_values(by='Count',ascending = False).head(10)

plt.figure(figsize = (12,6))

sns.barplot(data=top_10_groups, x='Count', y='Country', hue='Country')

plt.title('Top 10 Countries for the Largest Shared Profile Group')
plt.xlabel('Number of Respondents')
plt.ylabel('Country')

plt.show()
           
```
